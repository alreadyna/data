### jupyter notebook
- mode
    - 명령모드(esc): 셀 수정
    - 편집모드(enter): 셀 안의 내용 수정
- style
    - markdown(명령모드 > m): 셀 안의 설명 작성 시 사용
    - code(명령모드 > y): 파이썬 코드를 작성할 때 사용
- 단축키
    - 셀 실행: shift+enter
    - 셀 삭제: (명령모드)>x
    - 되돌리기: (명령모드)>z
    - 셀 생성: 위(명령모드)>a / 아래(명령모드)>b

```python
1+2
```

### magic command
- 셀 내부에서 특별하게 동작하는 커맨드
- %는 한줄의 masic command 를 동작
- %%: 셀 단위의 magic command 를 동작

- 주요 magic command
    - pwd: 현재 주피터노트북의 파일 경로
    - ls: 현재 디렉토리의 파일 리스트
    - whos: 현재 선언된 변수 출력
    - reset: 현재 선언된 변수 삭제

```python
%pwd
```

```python
a = 1
```

```python
whos
```

### shell command
- 주피터 노트북을 실행 쉘 환경의 명령을 사용
- 명령어 앞에 !를 붙여서 사용
- 주요 명령어
    - ls
    - cat
    - echo

```python
!echo python
```

```python
!ls
```
